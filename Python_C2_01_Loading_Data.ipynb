## **Python Function**

**Syntax for Writing Functions in Python**

The def keyword in Python is used to define a function.This function is a practical example of custom Python programming, useful in various fields including bioinformatics analysis or other contexts.

```python
#define our function named as seq_length
def seq_length(seq):
    counter = 0
    while seq[counter:]:
        counter += 1
    return counter #returns the sequence length
```


The provided code defines the function "seq_length(seq)" to calculate sequence length. It initializes a counter at 0 and iterates through the sequence while incrementing the counter. The function returns the counter, representing sequence length.

```python
#Execute our function, print length of given sequence
seq1 = "ATCTGGCAATGC"
print(seq_length(seq1))
```

The code assigns a sequence "ATCTGGCAATGC" to seq1. Then, it calls seq_length(seq1) to calculate and print the sequence length (12 characters). The seq_length function iterates through the sequence, counting elements and returning the length.
