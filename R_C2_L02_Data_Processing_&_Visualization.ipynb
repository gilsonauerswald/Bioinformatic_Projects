### **1. Load Data:**

Let’s practice some basic code that allows us to make different types of visualizations that summarize data properties and help us spot trends in our data visually:

```r
#data frame (df)
df = read.table('https://code.omicslogic.com/assets/datasets/15_genes/CellLines_15Genes.txt',row.names=1, header = TRUE, sep="\t")
head(df)
```

The code creates a DataFrame named df by reading data from the input file using read.table(). The read.table() function reads the file, treating the first column as row names, and using the header row for column names. The file is tab-separated ("\t"), and the correct file path should be provided.

### **2. Pre-process and Transform Data for Visualization:**

Data cleaning in R. First, let’s load the data and view it:

```r
#Remove a row, n is number of the column from a dataframe
newdata = df[-1,]
head(newdata)
dim(newdata)
```

The code creates a new DataFrame named newdata by excluding the first row from the original DataFrame data. The negative index -1 points to the first row, indicating its exclusion. The comma ensures all columns are retained in the new DataFrame.

```r
#Remove a column from dataframe
datanew = df[,-1]
head(datanew)
```

The code creates a new DataFrame named datanew by excluding the first column from the original DataFrame data. The negative index -1 refers to the first column, and the comma ensures all rows are retained in the new DataFrame.


After executing these lines of code, the datanew DataFrame will lack the first column, and the data DataFrame will still contain its original columns and rows but without any column names.

### **3. Remove NA Values**

```r
#Remove NAs from a dataset
Cleandata = na.exclude(df)
head(Cleandata)
```

The code creates a new dataset named Cleandata by applying the na.exclude() function to the original dataset data, excluding rows with missing values (NAs).

The function na.exclude() removes rows containing NAs from the dataset data, resulting in the Cleandata dataset with the same structure but without NA-containing rows.

### **4. Remove Zero or Low Values**

```r
#calculate row sums
cleandata_rowsum = rowSums(Cleandata)

#filter clean data based on rowsums
datanew <- Cleandata[cleandata_rowsum > 0, ]
head(datanew)
```

The R code above calculates the row sums of the dataset stored in Cleandata using the rowSums() function, assigning the result to the variable cleandata_rowsum. Subsequently, it filters the Cleandata based on the calculated row sums, creating a new dataset named datanew. The head(datanew) function is then used to display the first few rows of the filtered dataset for examination.

### **5. Log transformation:**

We can perform log scale transformation in R using log() function. Notably, here, we are adding +1 in data using datanew+1 to handle the zero values in the data, since log of zero is non-numeric.

```r
logdata = log(datanew+1)
boxplot(logdata)
```

logdata = log(datanew+1) performs a log transformation on the dataframe data, adding 1 before taking the logarithm. The transformed data is stored in logdata.

boxplot(logdata) creates a boxplot of logdata, visually depicting the distribution characteristics of the log-transformed values, including median, quartiles, and possible outliers.

### **6. Visualize Data**

We can visualize and create a boxplot using boxplot(), histogram using the hist() and heatmap using the heatmap() functions.

```r
#Load Data
df = read.table('https://code.omicslogic.com/assets/datasets/15_genes/CellLines_15Genes.txt',header = TRUE, sep='\t')

#Remove 1st column
df=df[,-1]

#Descriptive statistics
summary(df)

#Convert into matrix
df=as.matrix(df)

#Draw a boxplot
boxplot(df, col = rainbow(ncol(df)))

# Draw a Heatmap
heatmap(df)
```

The code snippet loads data from a URL into a dataframe named df, removes the first column, provides descriptive statistics, converts df into a matrix, creates a red-colored boxplot of column distributions, and generates a heatmap for data visualization and pattern recognition.

**Add title of the plot and change direction of legends on axis**

```r
#Draw a boxplot
boxplot(df,  col = rainbow(ncol(df)), las=2, main="Boxplot")
```

The code boxplot(df, col='red', las=2, main="Boxplot") creates a customized boxplot of values in dataframe df. The boxes are red, axis labels are rotated for readability, and the plot is titled "Boxplot." This visualization shows data distribution, including medians, quartiles, and potential outliers.
