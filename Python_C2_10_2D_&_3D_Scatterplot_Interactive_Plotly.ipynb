## **1. Load data and libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_table('https://raw.githubusercontent.com/PineBiotech/omicslogic/master/CellLines_15Genes_marked.txt',sep='\t',header=(0))
df
```

## **2. parse data for labels and save annotation**

```python
labels_list = df.loc[[0]]

#Make a list of sample names
sample_names = df.loc[0]
sample_names = list(df)[1:]

#Make unique list of labels
labels = np.unique(labels_list)

#delete "Group" label
labels = np.delete(labels, np.where(labels == 'Group'))

#Remove row "Group" at index 0 from data
data=df.drop([0])

#Remove id from axis
data=data.drop(['id'], axis=1)
data = data.transpose()

#Get list of all labels
flabels = labels_list.loc[0,:].values[1:]
```

```python
#flabels
```

## **3. Run PCA**

```python
from sklearn.preprocessing import StandardScaler
scaled = StandardScaler()
scaled.fit(data)
scaled_data = scaled.transform(data)

#Run Principal Component Analysis
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(scaled_data)
xpca = pca.transform(scaled_data)

#xpca
```

## **5. 2D scatterplot**

```python
# x and y given as array_like objects
import plotly.express as px

figPCA2D = px.scatter(xpca, x=0, y=1, color=flabels)
figPCA2D.show()
```

## **6. 3D scatterplot**

```python
# x and y given as array_like objects
import plotly.express as px


figPCA3D = px.scatter_3d(xpca, x=0, y=1, z=2, color=flabels)
figPCA3D.show()
```

## **7. Save PCA table**

```python
from pandas.core.frame import DataFrame

PCA_table = pd.DataFrame(xpca, columns=['PC1', 'PC2', 'PC3'], index=flabels)
#PCA_table
```

## **8. PCA table in plotly**

```python
import plotly.graph_objects as go

TablePCA = go.Figure(data=[go.Table(
    header=dict(values=list(['ID', 'PC1', 'PC2', 'PC3']),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[PCA_table.index, PCA_table.PC1, PCA_table.PC2, PCA_table.PC3],
               fill_color='lavender',
               align='left'))
])

TablePCA.show()
```

## **9. Plot PCA using Plotly**

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=2, cols=1,
    specs=[[{"type": "scatter"}],
           [{"type": "table"}]],
           shared_xaxes=True,
           vertical_spacing=0.05
)
for i in range(4):
  fig.add_trace(go.Scatter(figPCA2D.data[i]), row=1, col=1)

fig.add_trace(go.Table(
    header=dict(values=list(['ID', 'PC1', 'PC2', 'PC3']),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[PCA_table.index, PCA_table.PC1, PCA_table.PC2, PCA_table.PC3],
               fill_color='lavender',
               align='left')),
    row=2, col=1)

fig.update_layout(height=1000, width=800, title_text="PCA Results")
fig.show()
```

## **10. Plotly subplots**

```python
#from plotly.subplots import make_subplots
#import plotly.graph_objects as go

fig1 = make_subplots(
    rows=2, cols=1,
    specs=[[{"type": "scene"}],
           [{"type": "table"}]],
           shared_xaxes=True,
           vertical_spacing=0.05
)

for i in range(4):
  fig1.add_trace(go.Scatter3d(
       figPCA3D.data[i]), row=1, col=1)

fig1.add_trace(go.Table(
    header=dict(values=list(['ID', 'PC1', 'PC2', 'PC3']),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[PCA_table.index, PCA_table.PC1, PCA_table.PC2, PCA_table.PC3],
               fill_color='lavender',
               align='left')),
    row=2, col=1)

fig1.update_layout(height=1000, width=800, title_text="PCA Results")
fig1.show()
```
