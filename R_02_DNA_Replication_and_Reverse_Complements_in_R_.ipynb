# **DNA Replication and Reverse Complements in R**

## **Reverse Complements**

````r
if (!requireNamespace("BiocManager", quietly = TRUE))
    install.packages("BiocManager")
BiocManager::install("Biostrings")
library(Biostrings)
````

## **DNA Replication**

````r
library(Biostrings)

# Define a DNA sequence
dna_seq <- DNAString("ATCGGCTA")

# Replicate the DNA sequence
replicated_seq <- reverseComplement(dna_seq)

# Print the replicated sequence
print(replicated_seq)
````

## **Reverse Complement**

````r
library(Biostrings)

# Define a DNA sequence
dna_seq <- DNAString("ATCGGCTA")

# Obtain the reverse complement of the DNA sequence
rev_comp_seq <- reverseComplement(dna_seq)

# Print the reverse complement sequence
print(rev_comp_seq)
````

````r
#### load libraries
library(Biostrings)
#### Load sequence
seq <- DNAString("ATGCCGTGGTAAAGCCTTAAG") # or you can upload data from a fasta file with following step # read seq from a fasta file
seq <- readDNAStringSet("/content/gene.fna")
print(seq)
#### Compute the length of sequence
len <- length(seq)
print(len)
````

````r
#Now, Let’s Compute the frequency of each nucleotide in the sequence:
# Compute frequency of Each nucleotide and GC content in the sequence
c= letterFrequency(seq, "C", as.prob = T)
print(c)
g= letterFrequency(seq, "G", as.prob = T)
print(g)
a= letterFrequency(seq, "A", as.prob = T)
print(a)
t= letterFrequency(seq, "T", as.prob = T)
print(t)
#### Compute the gc content in a sequence ```
gc= letterFrequency(seq, "GC", as.prob = T)
print(gc)
#### view the sequence in a browser (optional). ﻿Uncomment (remove #) to view seq in your browser ```
# BrowseSeqs(seq, highlight=0) ```
#### Create reverse complement ```
rev = reverseComplement(seq)
print(rev)
#This gives us the following result if everything is working correctly. ``` CTTAAGGCTTTACCACGGCAT ```
#### You Can also write output in a file ```
writeXStringSet(rev, file="./output-Rev-comp.fa")
````
