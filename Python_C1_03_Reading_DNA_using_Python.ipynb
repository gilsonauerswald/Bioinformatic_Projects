## **2. Calculate nucleotide frequency**

```python
def calculate_nucleotide_frequency(dna_sequence):
    # Calculate the frequency of each nucleotide
    nucleotide_count = {'A': 0, 'T': 0, 'C': 0, 'G': 0}

    for nucleotide in dna_sequence:
        if nucleotide in nucleotide_count:
            nucleotide_count[nucleotide] += 1

    return nucleotide_count



# Example DNA sequence
dna_sequence = "ATGCTTCAGAAAGGTCTTACG"

# Compute frequencies and GC content
nucleotide_frequencies = calculate_nucleotide_frequency(dna_sequence)


# Display the results
print("Nucleotide Frequencies:", nucleotide_frequencies)
```

## **3. Calculate GC content**

```python
def calculate_gc_content(dna_sequence):
    # Calculate GC content
    gc_count = sum(dna_sequence.count(nucleotide) for nucleotide in 'GC')
    gc_content = (gc_count / len(dna_sequence)) * 100
    return gc_content

# Example DNA sequence
dna_sequence = "ATGCTTCAGAAAGGTCTTACG"

gc_content = calculate_gc_content(dna_sequence)

print("GC Content (%):", gc_content)
```

The given text is a collection of DNA sequences in FASTA format. Each sequence is represented by a header line starting with the ">" symbol, followed by the sequence itself.

Each sequence is a string of characters representing the DNA bases (A, T, G, C). The sequences can be used for various bioinformatics analyses or processing.

```python
#imports a library included in Python to give extra functions
import requests
from collections import Counter

# Function to download and parse a FASTA file from a URL, returning a dictionary {'label':'seq'}.
def download_and_parse_fasta(url):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Will raise an HTTPError if the HTTP request returned an unsuccessful status code

        FASTA_dict = {}
        FASTA_label = ""
        for line in response.text.split('\n'):
            line = line.rstrip()  # Removes spaces and newline chars each line
            if line.startswith(">"):  # Catches sequence label
                FASTA_label = line[1:]  # Set the label as 1st char and after
                FASTA_dict[FASTA_label] = ""  # Set sequence label in dict with empty string
            else:  # Must be sequence
                FASTA_dict[FASTA_label] += line
        return FASTA_dict
    except requests.RequestException as e:
        print(f"Error downloading file: {e}")
        return None
```


The code snippet imports the Counter class from the collections module and defines the download_and_parse_fasta function. This function reads a FASTA file, parses its sequences into a dictionary using labels as keys, and returns the dictionary. However, the code does not show the complete implementation, including base occurrence counting and result printing.

## **5. Function to count base occurrence in a FASTA file**

```python
# Function to count base occurrence and print results
def count_bases_from_url(url, sequence_label):
    # Run our download FASTA function and set returned FASTA_dict to FASTA variable
    FASTA = download_and_parse_fasta(url)
    if FASTA is None:
        return None

    # Select the sequence we want to look at from the dict
    sequence = FASTA.get(sequence_label)
    if sequence is None:
        print(f"Sequence label '{sequence_label}' not found in the FASTA file.")
        return None

    # Counts and returns the bases in the genome as a Counter object
    return Counter(sequence)

# Example usage (you'll need to replace 'your_fasta_file_url' with the actual URL and the sequence label)
url = 'https://raw.githubusercontent.com/pine-bio-support/DNA/master/seq1.fa'
sequence_label = 'seq1'
print(str(count_bases_from_url(url, sequence_label)))
```


The code snippet imports the Counter class from the collections module and defines the count_bases function. This function uses the download_and_parse_fasta function to retrieve a specific sequence, from a FASTA file, then employs the Counter class to count the occurrences of each base in the sequence. The result is converted to a string for output.
