## **1. R Function**

**Syntax for Writing Functions in R**

```r
func_name <- function (argument) {
statement
}
```

The R code defines a function named "func_name" that takes an argument. Inside the function, you can replace "statement" with the desired code for the function's purpose, adhering to R's function structure. This template allows you to create custom functions for specific tasks in bioinformatics analysis or other contexts.

Here, we are going to write a function seq_length to calculate the length of sequence:

```r
seq = "ATGCAAATTT"
seq_length <- function (seq) {
  length <- nchar(seq)
 print(length)
}
seq_length(seq)
```

The R function "seq_length" calculates and displays the length of a given sequence. It's structured with a parameter "seq" representing the sequence, and it calculates the length using "nchar" and prints it using "print." This function is useful for determining sequence lengths in bioinformatics analysis or similar tasks.

Execute function

```r
#Execute function, print the length of given sequence
seq_length("ATCTGGCAATGC")
```

You can execute the function to calculate and print the length (12 characters) of the given sequence "ATCTGGCAATGC". This is a practical illustration of how the function works in calculating sequence lengths.
